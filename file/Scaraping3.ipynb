```python
import time
import pandas as pd
import undetected_chromedriver as uc
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup




# สร้าง URL จากรายการ
url = f"https://shopee.co.th/3m_officialonlinestore"

options = webdriver.ChromeOptions()
options.headless = False

driver = uc.Chrome(options=options, use_subprocess=False)
driver.get(url)


try:
    # รอให้ป็อปอัพขึ้น
    time.sleep(3)
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="modal"]/div[1]/div[1]/div/div[3]/div[1]/button'))
    )

    # ค้นหา element ด้วย XPath และคลิก
    button = driver.find_element(By.XPATH, '//*[@id="modal"]/div[1]/div[1]/div/div[3]/div[1]/button')
    button.click()

    wait = WebDriverWait(driver, 10)
    button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="main"]/div/div[2]/div/div/div/div[2]/div/div[2]/div/div/div/a[2]')))
    button.click()
    
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'shopee-sort-bar'))
    )

    # ตัวอย่างการดึงข้อมูลที่มี class เป็น "Od5Z3o _2sze40 XwV4kb"
    # ตัวอย่างการดึงข้อมูลที่มี class เป็น "Od5Z3o _2sze40 XwV4kb"
# ตัวอย่างการดึงข้อมูลที่มี class เป็น "Od5Z3o _2sze40 XwV4kb"
    elements_class_Od5Z3o = driver.find_elements(By.CLASS_NAME, 'Od5Z3o')
    data_list_Od5Z3o = [element.text for element in elements_class_Od5Z3o]

    # ตัวอย่างการดึงข้อมูลที่มี class เป็น "MQbiLE"
    elements_class_MQbiLE = driver.find_elements(By.CLASS_NAME, 'MQbiLE')
    data_list_MQbiLE = [element.text for element in elements_class_MQbiLE]

    # ตัวอย่างการดึงข้อมูลที่มี class เป็น "_2VNMCr _8zpprh"
    elements_class_2VNMCr = driver.find_elements(By.CLASS_NAME, '_2VNMCr')
    data_list_2VNMCr = [element.text for element in elements_class_2VNMCr]

    # สร้าง DataFrame จากข้อมูลทั้งสามกลุ่ม
    df = pd.DataFrame({
        'ชื่อสินค้า': data_list_Od5Z3o,
        'ราคา': data_list_MQbiLE,
        'จำนวนที่ขายได้': data_list_2VNMCr
    })

    # แสดงตาราง DataFrame
    # ตัวอย่างการดึงข้อมูลที่มี class เป็น "shopee-button-no-outline"
    ##################
    elements_class_shopee_button = driver.find_elements(By.CLASS_NAME, 'shopee-button-no-outline')
    data_list_shopee_button = [element.text for element in elements_class_shopee_button]

    # แสดงข้อมูลในลิสต์
    print(data_list_shopee_button)
        

except Exception as e:
    print(f"An error occurred: {e}")
#finally:
    # ปิดเบราว์เซอร์
    #driver.quit()
df = df.drop_duplicates(subset=['ชื่อสินค้า', 'ราคา', 'จำนวนที่ขายได้'], keep='first')
df
```
