```python
# รายการของค่า list
list_values = [11044959,11044945, 11044964, 11044948, 11044952, 11044946, 11044955, 11044963, 11044960, 11044947,11044950, 11044953, 11044942, 11044944, 11044949, 11044954, 11044956, 11044951, 11044958, 11044943, 11044962, 11044961, 11044957]

# สร้าง DataFrame ว่างๆ เพื่อรวมผลลัพธ์
result_df = pd.DataFrame()

for value in list_values:
    # สร้าง URL จากรายการ
    url = f"https://shopee.co.th/mall/brands/{value}"

    # เปิดเบราว์เซอร์และเข้าไปที่ URL
    options = webdriver.ChromeOptions()
    options.headless = False

    driver = uc.Chrome(options=options)
    driver.get(url)

    try:
        # รอให้ป็อปอัพขึ้น
        WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="modal"]/div[1]/div[1]/div/div[3]/div[1]/button'))
        )

        # ค้นหา element ด้วย XPath และคลิก
        button = driver.find_element(By.XPATH, '//*[@id="modal"]/div[1]/div[1]/div/div[3]/div[1]/button')
        button.click()

        # รอให้หน้าเว็บโหลดเสร็จ
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, 'full-brand-list-item__brand-name')))

        # ค้นหา Element ที่มี class "full-brand-list-item__brand-name"
        brand_elements = driver.find_elements(By.CLASS_NAME, 'full-brand-list-item__brand-name')

        # ค้นหา Element ที่มี class "full-brand-list-item__brand-cover-image"
        brand_cover_elements = driver.find_elements(By.CLASS_NAME, 'full-brand-list-item__brand-cover-image')

        # ดึงข้อมูลและบันทึกลงในรายการ
        brand_list = [element.text for element in brand_elements]
        brand_list2 = [element.get_attribute("href") for element in brand_cover_elements]

        # สร้าง DataFrame จากรายการ
        temp_df = pd.DataFrame({'Brand Name': brand_list, 'Brand Link': brand_list2})

        # เพิ่มผลลัพธ์ใน DataFrame รวม
        result_df = pd.concat([result_df, temp_df], ignore_index=True)
        
        driver.close()
        
    except Exception as e:
        print(f"An error occurred: {e}")
        driver.close()

    finally:
        # ปิดเบราว์เซอร์
        
        driver.quit()

# พิมพ์ผลลัพธ์


driver.quit()

# Add these lines before executing the code block that caused the error
options = webdriver.ChromeOptions()
options.headless = False

driver = uc.Chrome(options=options)
print(result_df)
```

```python
import time
import pandas as pd
import undetected_chromedriver as uc
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

# Ensure that 'result_df' DataFrame is already loaded with the required data

options = webdriver.ChromeOptions()
options.headless = False  # Set headless to True for headless browsing

driver = uc.Chrome(options=options, use_subprocess=False)

# DataFrame to store scraped data from all brands
all_brands_df = pd.DataFrame()

for link in result_df['Brand Link'].iloc[0:5]:  # Iterate through each brand link
    for i in range(0, 100):  # Loop through pages
        driver.get(link + '?page=' + str(i))

        try:
            # Wait for the page to load and handle popup if present
            time.sleep(3)
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@id="modal"]/div[1]/div[1]/div/div[3]/div[1]/button'))
            )
            button = driver.find_element(By.XPATH, '//*[@id="modal"]/div[1]/div[1]/div/div[3]/div[1]/button')
            button.click()

            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, 'shopee-sort-bar'))
            )

            # Extract the required data
            elements_class_Od5Z3o = driver.find_elements(By.CLASS_NAME, 'Od5Z3o')
            elements_class_MQbiLE = driver.find_elements(By.CLASS_NAME, 'MQbiLE')
            elements_class_2VNMCr = driver.find_elements(By.CLASS_NAME, '_2VNMCr')

            # Create a temporary DataFrame for the current brand
            temp_df = pd.DataFrame({
                'ชื่อสินค้า': [element.text for element in elements_class_Od5Z3o],
                'ราคา': [element.text for element in elements_class_MQbiLE],
                'จำนวนที่ขายได้': [element.text for element in elements_class_2VNMCr]
            })

            # Append the temporary DataFrame to the main DataFrame
            all_brands_df = pd.concat([all_brands_df, temp_df], ignore_index=True)

        except TimeoutException:
            print(f"'shopee-sort-bar' not found on page {i} for {link}, moving to the next link.")
            break  # Break the inner loop to move to the next link

        except Exception as e:
            print(f"An error occurred on page {i} for {link}: {e}")
            continue  # Continue to the next page

# Drop duplicates if any
all_brands_df = all_brands_df.drop_duplicates(subset=['ชื่อสินค้า', 'ราคา', 'จำนวนที่ขายได้'], keep='first')

# Close the driver
driver.quit()

# Display or process the final DataFrame
print(all_brands_df)
```
